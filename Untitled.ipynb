# tsetse-odor-coding-analysis
This code analyzes the statistical relationships between different classes of olfactory receptor neurons in the sensing hairs (sensilla) of tsetse flies. The analyses here was used in the publication ["Odor coding in the antenna of the tsetse fly *Glossina morsitans*"](https://www.pnas.org/content/116/28/14300.full) by N. Soni, S. Chahda, and J. Carlson, published in May 2019 in *PNAS*. 

```python
import scipy as sp
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from scipy.stats import pearsonr, spearmanr
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
from sklearn import linear_model
```

## Load data from putative ab1, ab2, and ab3 tsetse sensilla

```python
filename = 'data/ab1.txt'
ab1 = sp.genfromtxt(filename)[:, :-1]
filename = 'data/ab2.txt'
ab2 = sp.genfromtxt(filename)[:, :-1]
filename = 'data/ab3.txt'
ab3 = sp.genfromtxt(filename)[:, :-1]

# Number of tested ORNs of each sensilla and number of odors
n1 = ab1.shape[0]
n2 = ab2.shape[0]
n3 = ab3.shape[0]
num_odors = ab1.shape[1]

# Statistics for each sensilla class 
ab1_mean = sp.mean(ab1, axis=0)
ab2_mean = sp.mean(ab2, axis=0)
ab3_mean = sp.mean(ab3, axis=0)
ab1_sem = sp.std(ab1, axis=0)/n1**0.5
ab2_sem = sp.std(ab2, axis=0)/n2**0.5
ab3_sem = sp.std(ab3, axis=0)/n3**0.5

# all_data has shape (ORNs, odors)
all_data = sp.vstack((ab1, ab2))
all_data = sp.vstack((all_data, ab3))
ab13 = sp.vstack((ab1, ab3))
splt = (n1, n1 + n2, n1 + n2 + n3)
```

## ORN firing rates for each sensilla class

```python
fig = plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.xlabel('Odor')
plt.ylabel('ab1 Firing rate (Hz)')
plt.bar(range(num_odors), height=ab1_mean, yerr=ab1_sem, lw=0, alpha=0.5)
plt.ylim(-10, 120)

plt.subplot(132)
plt.xlabel('Odor')
plt.ylabel('ab2 Firing rate (Hz)')
plt.bar(range(num_odors), height=ab2_mean, yerr=ab2_sem, lw=0, alpha=0.5)
plt.ylim(-10, 120)

plt.subplot(133)
plt.xlabel('Odor')
plt.ylabel('ab3 Firing rate (Hz)')
plt.bar(range(num_odors), height=ab3_mean, yerr=ab3_sem, lw=0, alpha=0.5)
plt.ylim(-10, 120)
plt.show()
```

## t-SNE clustering of ORN tuning curves

Each ORN is identified by its responses to the 15 tested odors. We reduce the 15D odor vector of responses to 2D using t-SNE nonlinear clustering. Each point thus represents a different ORN, with ORNs in putative ab1, ab2, and ab3 sensilla, colored in blue, red, and green, respectively.

```python
TSNE_func = TSNE(random_state=0, learning_rate=200)
reduced_data = TSNE_func.fit_transform(all_data)

fig = plt.figure()
fig.set_size_inches(4, 4)
plt.scatter(reduced_data[:splt[0], 0], reduced_data[:splt[0], 1], color='b')
plt.scatter(reduced_data[splt[0]: splt[1], 0], reduced_data[splt[0]: splt[1], 1], color='r')
plt.scatter(reduced_data[splt[1]: splt[2], 0], reduced_data[splt[1]: splt[2], 1], color='g')
plt.show()
```

## Correlation analysis
Plot sensilla class pairs against one another; Spearman rank coefficient of means to quantify. 

```python
ab1_mean = sp.mean(ab1, axis=0)
ab2_mean = sp.mean(ab2, axis=0)
ab3_mean = sp.mean(ab3, axis=0)

fig = plt.figure(figsize=(17, 4))
ax = plt.subplot(131)
plt.scatter(ab1_mean, ab3_mean, color='k')
plt.xticks(sp.arange(0, 200, 20))
plt.yticks(sp.arange(0, 200, 20))
ax.tick_params(axis='both', labelsize=12)
plt.xlabel('ab1 mean rates (Hz)', fontsize=15)
plt.ylabel('ab3 mean rates (Hz)', fontsize=15)
plt.xlim(0, 110)
plt.ylim(0, 110)

ax = plt.subplot(132)
plt.scatter(ab1_mean, ab2_mean, color='k')
plt.xticks(sp.arange(0, 200, 20))
plt.yticks(sp.arange(0, 200, 20))
ax.tick_params(axis='both', labelsize=12)
plt.xlabel('ab1 mean rates (Hz)', fontsize=15)
plt.ylabel('ab2 mean rates (Hz)', fontsize=15)
plt.xlim(-10, 70)
plt.ylim(-10, 70)

ax = plt.subplot(133)
plt.scatter(ab3_mean, ab2_mean, color='k')
plt.xticks(sp.arange(0, 200, 20))
plt.yticks(sp.arange(0, 200, 20))
plt.xlabel('ab3 mean rates (Hz)', fontsize=15)
plt.ylabel('ab2 mean rates (Hz)', fontsize=15)
ax.tick_params(axis='both', labelsize=12)
plt.xlim(-10, 110)
plt.ylim(-10, 110)
plt.show()

print (spearmanr(ab1_mean, ab3_mean))
print (spearmanr(ab1_mean, ab2_mean))
print (spearmanr(ab3_mean, ab2_mean))
```

## Correlation matrix analyses
For each sensillum class, plot the correlation matrices across firing responses to each odor

```python
C1 = sp.corrcoef(ab1.T)
C2 = sp.corrcoef(ab2.T)
C3 = sp.corrcoef(ab3.T)

fig = plt.figure(figsize=(17, 4))
plt.subplot(131)
plt.imshow(C1, vmin=-0.5, cmap=plt.cm.viridis)
plt.colorbar()

plt.subplot(132)
plt.imshow(C2, vmin=-0.5, cmap=plt.cm.viridis)
plt.colorbar()

plt.subplot(133)
plt.imshow(C3, vmin=-0.5, cmap=plt.cm.viridis)
plt.colorbar()
plt.show()
```

## Correlation of correlation matrices

This quantifies how correlated the off diagonal elements of the correlation matrices are to each other. Linear relationships with intercept 0 and near slope 1 are highly similar.

```python
C1_off_diag = C1[sp.triu(C1, k=1) != 0]
C2_off_diag = C2[sp.triu(C2, k=1) != 0]
C3_off_diag = C3[sp.triu(C3, k=1) != 0]

x = sp.arange(-1, 1.05, 0.1)

res = sp.polyfit(C1_off_diag, C3_off_diag, deg=1)
print (res)

fig = plt.figure(figsize=(15, 4))
ax = plt.subplot(131)
plt.scatter(C1_off_diag, C3_off_diag, color='k')
plt.plot(x, x*res[0] + res[1], color='r')
plt.xticks(sp.arange(-5, 5, 0.5))
plt.yticks(sp.arange(-5, 5, 0.5))
plt.xlabel('ab1 correlation coefficients', fontsize=15)
plt.ylabel('ab3 correlation coefficients', fontsize=15)
ax.tick_params(axis='both', labelsize=12)
plt.xlim(-0.5, 1.0)
plt.ylim(-0.5, 1.0)
plt.axhline(0, color='k', lw=0.5)
plt.axvline(0, color='k', lw=0.5)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.axis('equal')
ax = plt.subplot(132)
        
res = sp.polyfit(C2_off_diag, C1_off_diag, deg=1)
print (res)

ax = plt.subplot(132)
plt.scatter(C2_off_diag, C1_off_diag, color='k')
plt.plot(x, x*res[0] + res[1], color='r')
plt.xticks(sp.arange(-5, 5, 0.5))
plt.yticks(sp.arange(-5, 5, 0.5))
plt.xlabel('ab2 correlation coefficients', fontsize=15)
plt.ylabel('ab1 correlation coefficients', fontsize=15)
ax.tick_params(axis='both', labelsize=12)
plt.xlim(-0.5, 1.0)
plt.ylim(-0.5, 1.0)
plt.axhline(0, color='k', lw=0.5)
plt.axvline(0, color='k', lw=0.5)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.axis('equal')
plt.tight_layout()

res = sp.polyfit(C2_off_diag, C3_off_diag, deg=1)
print (res)

ax = plt.subplot(133)
plt.scatter(C2_off_diag, C3_off_diag, color='k')
plt.plot(x, x*res[0] + res[1], color='r')
plt.xticks(sp.arange(-5, 5, 0.5))
plt.yticks(sp.arange(-5, 5, 0.5))
plt.xlabel('ab2 correlation coefficients', fontsize=15)
plt.ylabel('ab3 correlation coefficients', fontsize=15)
ax.tick_params(axis='both', labelsize=12)
plt.xlim(-0.5, 1.0)
plt.ylim(-0.5, 1.0)
plt.axhline(0, color='k', lw=0.5)
plt.axvline(0, color='k', lw=0.5)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.axis('equal')
plt.tight_layout()
plt.show()
```
